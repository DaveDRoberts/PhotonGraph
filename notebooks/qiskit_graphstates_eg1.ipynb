```python
from qiskit import QuantumCircuit, execute, Aer
```

```python
import numpy as np
import sqlite3
import json

import photongraph as pg
```

# 7-Qubit STAR Graph State

```python
n = 7
circuit = QuantumCircuit(n)
```

```python
circuit.h(0)
circuit.cx(0,1)
circuit.cx(1,2)
circuit.cx(2,3)
circuit.cx(3,4)
circuit.cx(4,5)
circuit.cx(5,6)
circuit.h(1)
circuit.h(2)
circuit.h(3)
circuit.h(4)
circuit.h(5)
circuit.h(6)
```

```python
circuit.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
result = job.result()
outputstate = result.get_statevector(circuit, decimals=7)
```

```python
pg_sv = pg.StateVector(n,2, outputstate, qudit_order_rev=True)
```

```python
gs = pg.gs_from_sv(pg_sv)
```

```python
gs.draw()
```
