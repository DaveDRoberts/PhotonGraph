# Example: Qubit Hypergraph

```python
qudit_dim = 2
edge_dict = {(0, 5, 6):1, (0, 1, 2, 3, 4):1, (1, 7, 8):1, (5, 6, 7, 8):1}
g = pg.GraphState(edge_dict, qudit_dim)
g.draw()
```

# Example: Z Measurement (|1><1| projection) on Qubit Hypergraph State

```python
edge_dict = {(0, 1, 3): 1, (1, 2, 6): 1, (3, 4, 5): 1}
g = pg.GraphState(edge_dict, 2)
g.draw()
g.measure_Z(0, 1)
g.draw()
g.edges
```

# Example: EM on Ququint Graph State

```python
edge_dict = {(0, 1): 3, (1, 2): 2, (2, 3): 3, (0, 3): 1, (3,): 2}
g = pg.GraphState(edge_dict, 5)
g.draw()
g.EM(0, 3)
g.draw()
g.edges
```

# Example: Z Measurement (|2><2| projector) on Qutrit Hypergraph State

```python
edge_dict = {(0, 1): 2, (1, 2): 1, (2, 3): 1, (0, 1, 2): 1, (1, 2, 3): 2}
g = pg.GraphState(edge_dict, 3)
g.draw()
g.measure_Z(0, 2)
g.draw()
g.edges
```
